```python
# source: https://seaborn.pydata.org/examples/many_pairwise_correlations.html
def show_corr_matrix(df):
    # Compute the correlation matrix
    corr = df.corr(numeric_only = True)
    
    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))
    
    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))
    
    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    
    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
from sklearn.model_selection import train_test_split

X = df_num.drop(['salary'], axis = 1)
y = df_num['salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=666)
```

## Target encoding

```python
def target_encode(df, column, target):
    target_mean = df.groupby(column)[target].mean()
    df[column] = df[column].map(target_mean)
    return df

def target_encode_many(df, columns, target):
    for column in columns:
        df = target_encode(df, column, target)
    return df
```

```python
pd.options.mode.chained_assignment = None

target_enc_df = df.copy()
train_te = target_enc_df[:(int(len(df) * 0.85))]
test_te = target_enc_df[(int(len(df) * 0.85)):]

train_te = target_encode_many(train_te, ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation'], 'salary')
test_te = target_encode_many(test_te, ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation'], 'salary')

pd.options.mode.chained_assignment = 'warn'

train_te
```

```python
model = GradientBoostingRegressor(n_estimators=50, learning_rate=0.05, max_depth=3, random_state=666)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'Root Mean Squared Error: {rmse}')
print(f'R^2 Score: {r2}')
```
