```python
import numpy as np
import pandas as pd
from sklearn import datasets
from pandas import DataFrame
from matplotlib import pyplot as plt
```

### task 1

```python
iris = datasets.load_iris()
print(iris)
```

```python
# Смотрим на названия переменных
print(iris.feature_names)

# Смотрим на данные, выводим 10 первых строк:
print(iris.data[:10])

# Смотрим на целевую переменную:
print(iris.target_names)
print(iris.target)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)

# Добавляем столбец с целевой переменной:
df['target'] = iris.target

# Для наглядности добавляем столбец с сортами:
df['name'] = df.target.map(lambda x : iris.target_names[x])
df.head()
```

```python
# визуализации
import seaborn as sns

for feature_name in iris.feature_names:
    sns.displot(
        data = df,
        x = feature_name,
        hue = 'name',
        kde = True
    )
```

```python
sns.pairplot(df, hue = 'target')
```

### task 2

```python
feature1 = 'sepal length (cm)'
feature2 = 'petal length (cm)'

X = df[df.name != 'setosa'].filter(items = [feature1, feature2])
y = df[df.name != 'setosa'].target

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

xx, yy = np.meshgrid(np.linspace(X.loc[:, feature1].min() - 1, X.loc[:, feature1].max() + 1, 1000),
                     np.linspace(X.loc[:, feature2].min() - 1, X.loc[:, feature2].max() + 1, 1000))

print(xx, yy)

# Функция для визуализации разделяющей кривой
def plot_decision_boundary(clf, title):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X.loc[:, feature1], X.loc[:, feature2], c=y, cmap=plt.cm.coolwarm, marker='o', edgecolor='k')
    plt.title(title)

# Линейный дискриминант
lda = LinearDiscriminantAnalysis()
lda.fit(X.values, y)
plt.figure()
plot_decision_boundary(lda, 'Линейный дискриминант')

# Квадратичный дискриминант
qda = QuadraticDiscriminantAnalysis()
qda.fit(X.values, y)
plt.figure()
plot_decision_boundary(qda, 'Квадратичный дискриминант')

# Логистическая регрессия
lr = LogisticRegression()
lr.fit(X.values, y)
plt.figure()
plot_decision_boundary(lr, 'Логистическая регрессия')

# SVM с линейным ядром
svm_linear = SVC(kernel='linear')
svm_linear.fit(X.values, y)
plt.figure()
plot_decision_boundary(svm_linear, 'SVM (линейное ядро)')

# SVM с квадратичным ядром
svm_quad = SVC(kernel='poly', degree=2)
svm_quad.fit(X.values, y)
plt.figure()
plot_decision_boundary(svm_quad, 'SVM (квадратичное ядро)')

plt.show()
  
```
