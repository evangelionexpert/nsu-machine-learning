#### Задание 4
Найдите и подкрепите аргументами такое правило, что
1) все удовлетворяющие ему пассажиры выжили;
2) все удовлетворяющие ему пассажиры погибли.

Постарайтесь найти такие правила, чтобы под них попало как можно большее число пассажиров

```python
# 1) все удовлетворяющие условию пассажиры выжили
alive = titanic[(titanic.Fare > 263) | (titanic.Age > 74)]
assert(alive.equals(alive[alive.Survived == 1]))
print(len(alive))
# print(alive.describe())

# 2) все удовлетворяющие условию пассажиры погибли
not_alive = titanic[(titanic.Parch > 5) | (titanic.SibSp > 4)]
assert(not_alive.equals(not_alive[not_alive.Survived == 0]))
print(len(not_alive))

# в обоих случаях фильтруем по условиям, пограничным для класса пассажиров
```
